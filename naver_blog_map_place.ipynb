###  네이버 블로그 검색 API 사용하기

```python
encText = urllib.parse.quote("서울 핫플")
url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "&display=100&start=2"

request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
rescode
```

```python
naver_urls = []
postdate = []
titles = []

data = json.loads(response_body)['items']
print(data)

for row in data:
  if('blog.naver' in row['link']):
    naver_urls.append(row['link'])
    postdate.append(row['postdate'])
    title = row['title']

    # html 태그 제거
    pattern = '<[^>]*>' # <>로 감싸진 태그, [^>] = >를 제외한 모든 문자 
    title = re.sub(pattern=pattern, repl='', string=title)
    titles.append(title)    

print(f"naver_urls : {naver_urls}")
print(f"postdate : {postdate}")
print(f"titles : {titles}")
```

```python
blog_df = pd.DataFrame({'title' : titles, 'blog_url' : naver_urls, 'postdate' : postdate})
blog_df.head()
```

### 함수로 만들기

```python
def parse_html(data):
  naver_urls = []
  postdates = []
  titles = []

  for row in data:
    if('blog.naver' in row['link']):
      naver_urls.append(row['link'])
      postdates.append(row['postdate'])
      title = row['title']

      # html 태그 제거
      pattern = '<[^>]*>' # <>로 감싸진 태그, [^>] = >를 제외한 모든 문자 
      title = re.sub(pattern=pattern, repl='', string=title)
      titles.append(title)    
  
  return naver_urls, postdates, titles
```

```python
def update_df(blog_df, naver_urls, postdates, titles):
    new_data = pd.DataFrame({'title' : titles, 'blog_url' : naver_urls, 'postdate' : postdates})
    if blog_df is None or blog_df.empty:
        blog_df = new_data
    else:
        blog_df = pd.concat([blog_df, new_data], ignore_index=True)
    
    return blog_df
```

```python
def selenium_blog_crawling(blog_df, naver_urls):
    place_ids = []
    names = []
    addresses = []
    latitudes = []
    longitudes = []

    try:
        for url in naver_urls:
            print(url)
            driver.get(url)
            time.sleep(5)
            
            iframe = driver.find_element(By.ID, "mainFrame")
            driver.switch_to.frame(iframe)
            
            source = driver.page_source
            soup = BeautifulSoup(source, "html.parser")
            
            # 블로그에 있는 네이버 플레이스 링크만 가져오기 
            a_tag = soup.find("a", class_="se-map-info __se_link")
        
            if a_tag:
                data_linkdata = a_tag.get("data-linkdata")
                if data_linkdata:
                    data = json.loads(data_linkdata)
                    
                    # 필요한 데이터 추출
                    place_ids.append(data.get("placeId", ""))
                    names.append(data.get("name", ""))
                    addresses.append(data.get("address", ""))
                    latitudes.append(data.get("latitude", ""))
                    longitudes.append(data.get("longitude", ""))
            
            # 지도 바로가기가 없으면 빈 값 추가하기
            else:
                place_ids.append("")
                names.append("")
                addresses.append("")
                latitudes.append("")
                longitudes.append("")
                
        
        if 'place_id' in blog_df.columns:
            blog_df.loc[blog_df['place_id'].isna(), ['place_id', 'place_name', 'address', 'latitude', 'longitude']] = list(zip(place_ids, names, addresses, latitudes, longitudes))

        
        else: # place_id가 컬럼에 없으면 
            blog_df['place_id'] = place_ids
            blog_df['place_name'] = names
            blog_df['address'] = addresses
            blog_df['latitude'] = latitudes
            blog_df['longitude'] = longitudes
        
        return blog_df
            

    except:
        raise Exception("error")
```

```python
def crawling_naver_blog(blog_df, start, display, query):
  encText = urllib.parse.quote(query)
  url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "&start=" + str(start+1) + "&display=" + str(display)

  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request)
  rescode = response.getcode()

  if(rescode==200):
    response_body = response.read().decode('utf-8')
    data = json.loads(response_body)['items']
    naver_urls, postdates, titles = parse_html(data)
    blog_df = update_df(blog_df, naver_urls, postdates, titles)
    blog_df = selenium_blog_crawling(blog_df, naver_urls)
    return blog_df
  else:
    raise Exception("Error Code:" + rescode)
```

```python
blog_df = pd.DataFrame()

for i in range(250):
    blog_df = crawling_naver_blog(blog_df=blog_df, start=i, display=100, query='서울 핫플')
    time.sleep(5)
```
